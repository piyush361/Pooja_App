```python
# Import necessary modules
from PIL import Image
from IPython.display import display

class Task:
    def __init__(self, title, priority, image_path=None):
        self.title = title
        self.priority = priority
        self.image_path = image_path

    def display_task(self):
        print(f"Task: {self.title}, Priority: {self.priority}")
        if self.image_path:
            try:
                img = Image.open(self.image_path)
                display(img)
            except FileNotFoundError:
                print("Image not found!")
                
# Create tasks with associated images
task1 = Task("Color Traditional Art", "High", "/home/piyush/welc.jpg")
task2 = Task("Design 2", "Medium", "/home/piyush/a.jpg")

task1.display_task()
task2.display_task()
```
